```python
#1 - Bad
#2 - Average
#3 - Excellent
#This will be split in the following way. 
#1,2,3 --> Bad
#4,5,6,7 --> Average
#8,9,10 --> Excellent
#Create an empty list called Reviews
rating = []
for i in wine['quality']:
    if i >= 1 and i <= 3:
        rating.append('1')
    elif i >= 4 and i <= 7:
        rating.append('2')
    elif i >= 8 and i <= 10:
        rating.append('3')
wine['Rating'] = rating
```

```python
wine.columns
```

```python
wine['Rating'].unique()
```

```python
x = wine.iloc[:,:11]
y = wine['Rating']
```

```python
x.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#view the scaled features
print(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
#plot the graph to find the principal components
plt.figure(figsize=(10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
ax.grid(b=True, which='major', color='b', linestyle='-')
plt.show()
```

```python
#AS per the graph, we can see that 6 principal components attribute for 90% of variation in the data. 
#We shall pick the first 6 components for our prediction.
pca_new = PCA(n_components=6)
x_new = pca_new.fit_transform(x)
```

```python
print(x_new)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = 0.25)
```

```python
print(x_train.shape)
print(y_train)
print(x_test.shape)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)
```

```python
#print confusion matrix and accuracy score
lr_conf_matrix = confusion_matrix(y_test, lr_predict)
lr_acc_score = accuracy_score(y_test, lr_predict)
print(lr_conf_matrix)
print(lr_acc_score*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt_predict = dt.predict(x_test)
```

```python
#print confusion matrix and accuracy score
dt_conf_matrix = confusion_matrix(y_test, dt_predict)
dt_acc_score = accuracy_score(y_test, dt_predict)
print(dt_conf_matrix)
print(dt_acc_score*100)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
nb_predict=nb.predict(x_test)
```

```python
#print confusion matrix and accuracy score
nb_conf_matrix = confusion_matrix(y_test, nb_predict)
nb_acc_score = accuracy_score(y_test, nb_predict)
print(nb_conf_matrix)
print(nb_acc_score*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_predict=rf.predict(x_test)
```

```python
#print confusion matrix and accuracy score
rf_conf_matrix = confusion_matrix(y_test, rf_predict)
rf_acc_score = accuracy_score(y_test, rf_predict)
print(rf_conf_matrix)
print(rf_acc_score*100)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import validation_curve
```

```python
svc = SVC()
svc.fit(x_train, y_train)
svc_predict=rf.predict(x_test)
```

```python
#print confusion matrix and accuracy score
svc_conf_matrix = confusion_matrix(y_test, svc_predict)
svc_acc_score = accuracy_score(y_test, svc_predict)
print(rf_conf_matrix)
print(rf_acc_score*100)
```

```python
param_range = np.arange(0.1, 50, 1)

train_scores, test_scores = validation_curve(SVC(),x_train, y_train,  param_name="C", param_range=param_range,  scoring="accuracy",n_jobs=2,)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)

test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
plt.subplots(1, figsize=(7,7))
plt.plot(param_range, train_mean, label="Training score", color="purple")
plt.plot(param_range, test_mean, label="Cross-validation score", color="green")

plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="blue")
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="red")
 
plt.title("Validation Curve With SVC")
plt.ylabel("Accuracy Score")
plt.xlabel("C")
plt.tight_layout()
plt.legend(loc="best")
plt.show()
```

```python
#Gridsearch code
from sklearn.model_selection import GridSearchCV
model = SVC()

tuning_parameters = {
                'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                'kernel' : ('linear', 'poly', 'rbf', 'sigmoid'),
                'max_iter' : [1, 10, 100, 1000]

}

CV = 5
VERBOSE = 0

# Run GridSearchCV for the model
grid_tuned = GridSearchCV(model,
                          tuning_parameters,
                          cv=CV,
                          scoring='accuracy',
                          verbose=VERBOSE,
                          n_jobs=-1
                         )

sv = grid_tuned.fit(x_train, y_train)

# Report result
print('Best score:', sv.best_score_*100)
print('Best settings:', sv.best_estimator_)
print(sv.scorer_)
print('OK(grid-search)')
```
